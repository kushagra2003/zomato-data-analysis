```python
import pandas as pd

# Try reading with encoding='latin1' or 'ISO-8859-1'
df = pd.read_csv('zomato.csv', encoding='latin1')

# Preview first few rows
df.head()
```

```python
# Check the shape of the dataset
print("Dataset Shape:", df.shape)

# See the column names
print("\nColumns:\n", df.columns)

# Check data types and non-null counts
df.info()

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Drop unnecessary columns if they exist
columns_to_drop = ['url', 'address', 'phone', 'dish_liked', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)', 'switch_to_order_menu']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Drop rows with any missing values
df.dropna(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(df.columns)
```

```python
city_counts = df['City'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=city_counts.values, y=city_counts.index, palette='viridis')
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Ensure no missing values in 'Cuisines' column
df = df[df['Cuisines'].notnull()]

# Break comma-separated cuisines into a list
cuisine_series = df['Cuisines'].str.split(', ')
cuisines_flat = [c for sublist in cuisine_series for c in sublist]
top_cuisines = Counter(cuisines_flat).most_common(10)

# Convert to DataFrame
cuisine_df = pd.DataFrame(top_cuisines, columns=['Cuisine', 'Count'])

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Count', y='Cuisine', data=cuisine_df, palette='rocket')
plt.title("Top 10 Cuisines Across India")
plt.tight_layout()
plt.show()
```

```python
# Convert to numeric if needed
df['Average Cost for two'] = pd.to_numeric(df['Average Cost for two'], errors='coerce')

# Drop rows where cost is NaN
df.dropna(subset=['Average Cost for two'], inplace=True)

# Group and plot
avg_cost = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_cost.values, y=avg_cost.index, palette='mako')
plt.title("Top 10 Cities by Avg Cost for Two")
plt.xlabel("Average Cost (â‚¹)")
plt.tight_layout()
plt.show()
```

```python
delivery_counts = df['Has Online delivery'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=delivery_counts.index, y=delivery_counts.values, palette='Set2')
plt.title("Online Delivery Availability")
plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()
```
